```python
import numpy as np
print(np.__version__)
```

```python
# Создание массивов из списков
a = np.array([1, 2, 3, 4, 5])
b = np.array([[1, 2, 3],
              [4, 5, 6]])

print("a:", a)
print("b:\n", b)

# zeros, ones, arange
zeros_arr = np.zeros((2, 3))
ones_arr = np.ones(4)
range_arr = np.arange(0, 10, 2)

print("zeros_arr:\n", zeros_arr)
print("ones_arr:", ones_arr)
print("range_arr:", range_arr)
```

```python
print("range_arr shape:", range_arr.shape)
print("range_arr dtype:", range_arr.dtype)

print("b shape:", b.shape)
print("b dtype:", b.dtype)
```

```python
print("Первый элемент range_arr:", range_arr[0])
print("Последний элемент range_arr:", range_arr[-1])
print("Срез range_arr[1:4]:", range_arr[1:4])

print("Элемент b[0, 1]:", b[0, 1])
print("Вторая строка b[1, :]:", b[1, :])
print("Первый столбец b[:, 0]:", b[:, 0])
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 30, 40])

print("x + y:", x + y)
print("x * y:", x * y)
print("mean(x):", x.mean())
print("max(y):", y.max())
```
